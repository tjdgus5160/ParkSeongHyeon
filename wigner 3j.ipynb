# $5S_{1/2} (F=3)  -> 5P_{1/2} (F'=3)$

```python
# 5S1/2 (F=3) -> 5P1/2 (F'=3), pi = 0

import sympy as sp

#   
L_e = 1
L_g = 0
S = sp.Rational(1, 2)  # SymPy  
J_e = sp.Rational(1, 2)
J_g = sp.Rational(1, 2)
F_g = 3
F_e = 3
I = sp.Rational(5, 2)
q = 0

#  1: (-1) ** (2*F_e + I + 2*J_g + L_e + S - 3)
m_F_e = 3
result1_1 = (-1) ** (2 * F_e + I + 2 * J_g + L_e + S - m_F_e)
m_F_e = 2
result1_2 = (-1) ** (2 * F_e + I + 2 * J_g + L_e + S - m_F_e)
m_F_e = 1
result1_3 = (-1) ** (2 * F_e + I + 2 * J_g + L_e + S - m_F_e)
m_F_e = 0
result1_4 = (-1) ** (2 * F_e + I + 2 * J_g + L_e + S - m_F_e)
m_F_e = -1
result1_5 = (-1) ** (2 * F_e + I + 2 * J_g + L_e + S - m_F_e)
m_F_e = -2
result1_6 = (-1) ** (2 * F_e + I + 2 * J_g + L_e + S - m_F_e)
m_F_e = -3
result1_7 = (-1) ** (2 * F_e + I + 2 * J_g + L_e + S - m_F_e)
print(f"(-1) ** (2*F_e + I + 2*J_g + L_e + S - m_F_e): {result1_1}")

#  2: ( (2*L_e + 1) * (2*F_g + 1) * (2*F_e + 1) * (2*J_g + 1) * (2*J_e + 1) ) ** (1/2)
result2 = sp.sqrt(
    (2 * L_e + 1) *
    (2 * F_g + 1) *
    (2 * F_e + 1) *
    (2 * J_g + 1) *
    (2 * J_e + 1)
)
print(f"( (2*L_e + 1) * (2*F_g + 1) * (2*F_e + 1) * (2*J_g + 1) * (2*J_e + 1) ) ** (1/2): {result2}")

# Wigner 6j symbol
result_6j_1 = wigner_6j(L_e, 1, L_g, J_g, S, J_e)
print(f"wigner_6j(L_e, 1, L_g, J_g, S, J_e): {result_6j_1}")

result_6j_2 = wigner_6j(J_e, 1, J_g, F_g, I, F_e)
print(f"wigner_6j(J_e, 1, J_g, F_g, I, F_e): {result_6j_2}")

# Wigner 3j symbol
print(f"wigner_3j(F_e, 1, F_g, -m_F_e, q, m_F_g), m_F_e = 3, m_F_g = 3: {wigner_3j(F_e,1,F_g,-3,q,3)}")
print(f"wigner_3j(F_e, 1, F_g, -m_F_e, q, m_F_g), m_F_e = 2, m_F_g = 2: {wigner_3j(F_e,1,F_g,-2,q,2)}")
print(f"wigner_3j(F_e, 1, F_g, -m_F_e, q, m_F_g), m_F_e = 1, m_F_g = 1: {wigner_3j(F_e,1,F_g,-1,q,1)}")
print(f"wigner_3j(F_e, 1, F_g, -m_F_e, q, m_F_g), m_F_e = 0, m_F_g = 0: {wigner_3j(F_e,1,F_g,0,q,0)}")
print(f"wigner_3j(F_e, 1, F_g, -m_F_e, q, m_F_g), m_F_e = -1, m_F_g = -1: {wigner_3j(F_e,1,F_g,1,q,-1)}")
print(f"wigner_3j(F_e, 1, F_g, -m_F_e, q, m_F_g), m_F_e = -2, m_F_g = -2: {wigner_3j(F_e,1,F_g,2,q,-2)}")
print(f"wigner_3j(F_e, 1, F_g, -m_F_e, q, m_F_g), m_F_e = -3, m_F_g = -3: {wigner_3j(F_e,1,F_g,3,q,-3)}")

# final result
print(f"C(F_e=3, m_F_e=3, F_g=3, m_F_g=3): {result1_1 * result2 * result_6j_1 * result_6j_2 * wigner_3j(F_e,1,F_g,-3,q,3)}")
print(f"C(F_e=3, m_F_e=2, F_g=3, m_F_g=2): {result1_2 * result2 * result_6j_1 * result_6j_2 * wigner_3j(F_e,1,F_g,-2,q,2)}")
print(f"C(F_e=3, m_F_e=1, F_g=3, m_F_g=1): {result1_3 * result2 * result_6j_1 * result_6j_2 * wigner_3j(F_e,1,F_g,-1,q,1)}")
print(f"C(F_e=3, m_F_e=0, F_g=3, m_F_g=0): {result1_4 * result2 * result_6j_1 * result_6j_2 * wigner_3j(F_e,1,F_g,0,q,0)}")
print(f"C(F_e=3, m_F_e=-1, F_g=3, m_F_g=-1): {result1_5 * result2 * result_6j_1 * result_6j_2 * wigner_3j(F_e,1,F_g,1,q,-1)}")
print(f"C(F_e=3, m_F_e=-2, F_g=3, m_F_g=-2): {result1_6 * result2 * result_6j_1 * result_6j_2 * wigner_3j(F_e,1,F_g,2,q,-2)}")
print(f"C(F_e=3, m_F_e=-3, F_g=3, m_F_g=-3): {result1_7 * result2 * result_6j_1 * result_6j_2 * wigner_3j(F_e,1,F_g,3,q,-3)}")
```
