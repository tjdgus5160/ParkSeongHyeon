```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
k = 14.4 # 쿨롱 상수 (eV nm)
E = 1e-1 # 전기장의 세기 (V/nm)

z = np.arange(-80, 80, 1) # z축의 범위와 간격 (nm)

V = -k/abs(z) + E*z # potential의 값 (eV)

plt.plot(z, V) # 그래프 그리기
plt.xlabel('z (nm)') # x축 라벨
plt.ylabel('V (eV)') # y축 라벨
plt.title('Potential of an atom in a uniform electric field') # 제목
plt.savefig("saddle point.png", dpi=500)
plt.show() # 그래프 보여주기

```

```python
k = 14.4 # 쿨롱 상수 (eV nm)
E = 1e-1 # 전기장의 세기 (V/nm)

z = np.arange(-80, 80, 1) # z축의 범위와 간격 (nm)

V = -k/abs(z)  # potential의 값 (eV)

plt.plot(z, V) # 그래프 그리기
plt.xlabel('z (nm)') # x축 라벨
plt.ylabel('V (eV)') # y축 라벨
plt.title('Potential of an atom') # 제목
plt.savefig("normal point.png", dpi=500)
plt.show() # 그래프 보여주기

```
