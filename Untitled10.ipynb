```python
import numpy as np
import matplotlib.pyplot as plt

# 볼츠만 분포 정의
def boltzmann_distribution(x, beta):
  return np.exp(-beta * x)

# x축과 y축 범위 설정
x = np.linspace(-10, 10, 100)
y = boltzmann_distribution(x, 1)

# 그래프 그리기
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Boltzmann Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 볼츠만 분포 함수 정의
def boltzmann_distribution(E, T):
    """
    E: 열적 에너지 (에너지 레벨)
    T: 온도 (켈빈)
    """
    k = 1.38e-23  # 볼츠만 상수 (J/K)
    return np.exp(-E / (k * T))

# 온도와 에너지 범위 설정
temperature = 300  # 온도 (300K = 27°C)
energy_range = np.linspace(0, 5e-21, 100)  # 에너지 범위 (0에서 5e-21 J 사이의 100개 데이터)

# 볼츠만 분포를 계산
distribution = boltzmann_distribution(energy_range, temperature)

# 그래프 그리기
plt.plot(energy_range, distribution, label=f'Temperature = {temperature}K')
plt.title('Boltzmann Distribution')
plt.xlabel('Energy (Joules)')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)

# 그래프 표시
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 루비듐(Rb) 원자의 795 nm 파장에서의 흡수 확률 계산 함수
def absorption_probability(wavelength):
    # 루비듐 원자의 특성 및 레이저 파장을 고려한 흡수 확률 계산
    # 이 예제에서는 임의의 공식을 사용합니다.
    # 실제로는 루비듐의 원자 스펙트럼 및 레이저 특성을 고려하는 복잡한 모델이 필요합니다.
    return np.sin(2 * np.pi * wavelength / 795)**2

# 파장(λ) 범위 설정 (예: 700 nm에서 900 nm)
wavelength_range = np.linspace(700, 900, 500)  # 700nm에서 900nm 범위 내에서 500개 데이터

# 흡수 확률 계산
absorption_probabilities = absorption_probability(wavelength_range)

# 그래프 그리기
plt.plot(wavelength_range, absorption_probabilities)
plt.title('Absorption Probability vs. Wavelength for Rb Atom')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorption Probability')
plt.grid(True)

# 그래프 표시
plt.show()
```
