```python
import numpy as np
import matplotlib.pyplot as plt

# 주요 변수 설정
frequency = 100e3  # 주파수 100kHz
period = 1 / frequency
time_long_wave = np.linspace(0, 3 * period, 3000)  # 3주기 동안의 시간 배열
tau = 65e-9  # 감쇠 시간 상수 65 ns

# 주기적인 펄스 신호를 생성하는 함수 정의
def generate_periodic_pulses(time_array, frequency, tau):
    period = 1 / frequency
    pulses = np.zeros_like(time_array)
    pulse_duration = period / 2  # 펄스의 길이는 주기의 절반
    for t in np.arange(0, time_array[-1], period):
        start_index = np.searchsorted(time_array, t)
        end_index = np.searchsorted(time_array, t + pulse_duration)
        pulses[start_index:end_index] = np.exp(-(time_array[start_index:end_index] - t) / tau)
    return pulses

# 주기적인 펄스 신호 생성
periodic_pulses = generate_periodic_pulses(time_long_wave, frequency, tau)

# 100kHz 참조 신호 생성 (sine wave)
reference_sine = np.sin(2 * np.pi * frequency * time_long_wave)

# Mixing 과정
mixed_signal = periodic_pulses * reference_sine

# Low-pass filter 정의 및 적용
def low_pass_filter(signal, window_size=50):
    return np.convolve(signal, np.ones(window_size)/window_size, mode='same')

filtered_signal = low_pass_filter(mixed_signal, 50)  # window_size를 조정하여 필터링

# 결과를 그래프로 표시
plt.figure(figsize=(12, 6))
plt.plot(time_long_wave * 1e6, periodic_pulses, label='Periodic Pulses', color='blue')
plt.plot(time_long_wave * 1e6, reference_sine, label='Reference Sine Wave', color='green')
plt.plot(time_long_wave * 1e6, mixed_signal, label='Mixed Signal', color='red')
plt.plot(time_long_wave * 1e6, filtered_signal, label='Filtered Signal', color='purple', linestyle='--')
plt.title('Periodic Pulses with Mixing and Filtering')
plt.xlabel('Time (microseconds)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square, sawtooth, butter, filtfilt  # LPF functions and waves

# Constants based on SR850 Specs
hbar = 1.054571817e-34  # Planck constant (J·s)
c = 3e8  # light speed (m/s)
d_eg = 1.46513e-29  # excited state dipole moment (C·m)
N0 = 3.93e23  # atom number density (m^-3)
V = 3.64e-5  # volume (m^3)
epsilon_0 = 8.854e-12  # vacuum permittivity (F/m)
T = 300  # temperature (K)
k_B = 1.38e-23  # Boltzmann constant (J/K)
m = 85.47 * 1.660539066e-27  # Rubidium mass (kg)

alpha = 72.77077e6  # polarizability of the Rydberg state (MHz / (V^2/cm^2))
E_amplitude = 0.3  # external electric field amplitude (V/cm)
E_frequency = 100e3  # 100 kHz square wave frequency

# Time-dependent variable
f_scan = 4  # Scan frequency in Hz (4 Hz period)
t_vals = np.linspace(0, 5 / f_scan, 40000)  # Time steps over one period of 4 Hz

# Generate 100 kHz square wave for EIT modulation
E_signal = E_amplitude * square(2 * np.pi * E_frequency * t_vals)

# Generate 100 kHz sine wave for PLL reference (from lock-in amplifier)
PLL_reference = np.sin(2 * np.pi * E_frequency * t_vals)

# Amplitude of Delta_c modulation (detuning of the coupling laser)
Delta_c_amplitude = (2 * np.pi * (-1/2) * alpha * E_signal ** 2)  # Stark effect modulation
Delta_c_triangle = sawtooth(2 * np.pi * f_scan * t_vals, 0.5) * 300e6  # Triangle wave modulation

# Initialize array to store imaginary part of chi over time
chi_imag_vals = np.zeros(len(t_vals))

# Fixed probe laser detuning
Delta_p = 0  # Probe detuning set to zero (Δ_p = 0)

# Coupling laser parameters
Omega_c = 2 * np.pi * 1.62e6  # constant coupling laser Rabi frequency (Hz)
gamma_p = 1e6  # probe laser linewidth
gamma_c = 1e6  # coupling laser linewidth
gamma_rel = (gamma_p + gamma_c) / 2
Gamma_e = 2 * np.pi * 5.75e6  # Decay rate of the excited state (Hz)
Gamma_r = 1 / 7.06e-05  # Decay rate of the Rydberg state (Hz)

# Loop over time steps to compute chi for each time step
for t_idx, t in enumerate(t_vals):
    Delta_c_t = Delta_c_amplitude[t_idx] + Delta_c_triangle[t_idx]  # Detuning modulated by triangle wave and Stark effect

    # Calculate susceptibility (chi) for Delta_p = 0 and modulated Delta_c
    denominator = (1/2) * Gamma_e + gamma_p - 1j * Delta_p + (Omega_c ** 2 / 4) / (
                (1/2) * Gamma_r + gamma_rel - 1j * (Delta_p + Delta_c_t))
    chi = -(1j * N0 * d_eg ** 2) / (epsilon_0 * hbar * denominator)
    
    # Store the imaginary part of chi (EIT peak-related component)
    chi_imag_vals[t_idx] = np.imag(chi)

# Differentiate the EIT signal to find the slope
chi_derivative = np.diff(chi_imag_vals) / np.diff(t_vals)

# Find the first point where the derivative crosses zero (side of fringe)
zero_crossings = np.where(np.diff(np.sign(chi_derivative)))[0]
if len(zero_crossings) > 0:
    stabilization_point = zero_crossings[0]  # First zero-crossing

    # Fix the laser detuning at the stabilization point
    stabilization_Delta_c = Delta_c_amplitude[stabilization_point] + Delta_c_triangle[stabilization_point]

    # After stabilization, fix Delta_c, but allow the E_signal to modulate the amplitude
    for t_idx in range(stabilization_point, len(t_vals)):
        Delta_c_t = stabilization_Delta_c  # Stabilize the laser detuning at side of fringe

        # Recalculate the susceptibility with the stabilized Delta_c, but allow E_signal to modulate the amplitude
        denominator = (1/2) * Gamma_e + gamma_p - 1j * Delta_p + (Omega_c ** 2 / 4) / (
                    (1/2) * Gamma_r + gamma_rel - 1j * (Delta_p + Delta_c_t))
        chi = -(1j * N0 * d_eg ** 2 * E_signal[t_idx]) / (epsilon_0 * hbar * denominator)  # Modulate the amplitude with E_signal

        # Update the imaginary part of chi during stabilization with amplitude modulation
        chi_imag_vals[t_idx] = np.imag(chi)

# ---------------- LOCK-IN AMPLIFIER SIMULATION ----------------

# 1. Mixer: Multiply the EIT signal (chi_imag_vals) with the reference sine wave (PLL_reference)
mixer_output = chi_imag_vals * PLL_reference

# 2. Low-Pass Filter (LPF) design: Apply LPF to remove high-frequency components
def butter_lowpass_filter(data, cutoff, fs, order=4):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y

# Low-pass filter specs based on SR850
cutoff_frequency = 500  # Cutoff frequency of the LPF (e.g., 500 Hz)
filter_order = 4  # Use 24 dB/octave roll-off
sampling_rate = len(t_vals) / (t_vals[-1] - t_vals[0])  # Sampling rate based on time steps

# Apply the low-pass filter to the mixer output
lpf_output = butter_lowpass_filter(mixer_output, cutoff_frequency, sampling_rate, filter_order)

# ---------------- PLOTTING RESULTS ----------------

# Plot the original EIT signal (imaginary part of chi)
plt.figure(figsize=(10, 6))
plt.plot(t_vals, chi_imag_vals, label='Imaginary part of χ (EIT peak)', color='red')
if len(zero_crossings) > 0:
    plt.axvline(t_vals[stabilization_point], color='black', linestyle='--', label='Stabilization Point')
plt.xlabel('Time (s)')
plt.ylabel('Imaginary part of χ')
plt.title('EIT Peak Evolution over Time (Δ_p = 0, Stabilization at Side of Fringe with Amplitude Modulation)')
plt.grid(True)
plt.legend()
plt.show()

# Plot the triangle wave signal to visualize the detuning and stabilization
plt.figure(figsize=(10, 6))
plt.plot(t_vals, Delta_c_triangle, label='4 Hz Triangle Wave (Detuning Modulation)', color='blue')
if len(zero_crossings) > 0:
    plt.axvline(t_vals[stabilization_point], color='black', linestyle='--', label='Stabilization Point')
plt.xlabel('Time (s)')
plt.ylabel('Delta_c (Hz)')
plt.title('Triangle Wave Modulation of Detuning (Δ_c) with Stabilization')
plt.grid(True)
plt.legend()
plt.show()

# Plot the mixer output (EIT signal * PLL sine wave)
plt.figure(figsize=(10, 6))
plt.plot(t_vals, mixer_output, label='Mixer Output (EIT * PLL Reference)', color='green')
if len(zero_crossings) > 0:
    plt.axvline(t_vals[stabilization_point], color='black', linestyle='--', label='Stabilization Point')
plt.xlabel('Time (s)')
plt.ylabel('Mixer Output Signal')
plt.title('Mixer Output (EIT signal * 100kHz PLL Sine Wave)')
plt.grid(True)
plt.legend()
plt.show()

# Plot the LPF output (lock-in amplifier output)
plt.figure(figsize=(10, 6))
plt.plot(t_vals, lpf_output, label='Lock-in Amplifier Output (LPF Applied)', color='purple')
if len(zero_crossings) > 0:
    plt.axvline(t_vals[stabilization_point], color='black', linestyle='--', label='Stabilization Point')
plt.xlabel('Time (s)')
plt.ylabel('Lock-in Amplifier Output Signal')
plt.title('Lock-in Amplifier Output (After LPF) with Stabilization')
plt.grid(True)
plt.legend()
plt.show()
```
